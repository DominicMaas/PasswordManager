```python
import matplotlib.pyplot as plt
from matplotlib.pyplot import figure
import operator

max_val = data[max(data.items(), key=operator.itemgetter(1))[0]]
min_val = data[min(data.items(), key=operator.itemgetter(1))[0]]

print(max_val)
print(min_val)

plt.figure(figsize=(14, 5), dpi=100)
plt.bar(*zip(*data.items()))
plt.ylim(bottom=min_val, top=max_val)

plt.show()
```
